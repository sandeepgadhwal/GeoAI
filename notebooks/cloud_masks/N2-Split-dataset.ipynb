```python
from pathlib import Path
import numpy as np
import shutil
```

```python
data_path = Path("/home/sandeep/workspace/competitions/MaskingCloudsinSatelliteImageries/data")
```

```python
file_list = np.array(list((data_path / 'train_true_color').glob('*.tif')))
file_list
```

```python
np.random.shuffle(file_list)
```

```python
val_split = 0.2
val_nos = int(val_split*len(file_list))
val_split, train_split = file_list[:val_nos], file_list[val_nos:]
```

```python
train_folder = data_path / 'split' / 'train'
val_folder = data_path / 'split' / 'val'
for folder, files in [
    [val_folder, val_split],
    [train_folder, train_split]
]:
    if folder.exists():
        shutil.rmtree(folder)
    folder.mkdir()
    img_folder = folder / 'image'
    mask_folder = folder / 'mask'
    img_folder.mkdir()
    mask_folder.mkdir()
    for file in files:
        shutil.copy(file, img_folder / file.name)
        idx = file.stem.split("_")[-1]
        label_file = file.parent.parent / "train_mask" / f"train_mask_{idx}.tif"
        shutil.copy(label_file, mask_folder / label_file.name)

    print(img_folder, len(list(img_folder.glob('*.tif'))))
    print(mask_folder, len(list(mask_folder.glob('*.tif'))))
```
