```python
from geoai.data.human_settlements import HumanSettlementsDataset
from collections import defaultdict
import numpy as np
```

```python
ds = HumanSettlementsDataset('/home/sandeep/workspace/data/human-settlements/', 2048)
```

```python
store = defaultdict(list)

# First pass
for i, batch in enumerate(ds):
    print(f"-- Progress Pass 1 | {i}/{len(ds)} {(i*100/len(ds)):0.2f} %", end='\r')
    arr = batch['img']
    arr_view = arr.reshape((-1, arr.shape[-1]))
    store['max'].append(arr_view.max(axis=0))
    store['min'].append(arr_view.min(axis=0))
    store['mean'].append(arr_view.mean(axis=0))

store['max'] = np.stack(store['max']).max(axis=0)
store['min'] = np.stack(store['min']).min(axis=0)
store['mean'] = np.stack(store['mean']).mean(axis=0)
```

```python

# Second pass
variance = []
for i, batch in enumerate(ds):
    print(f"-- Progress Pass 2 | {i}/{len(ds)} {(i*100/len(ds)):0.2f} %", end='\r')
    arr = batch['img']
    arr_view = arr.reshape((-1, arr.shape[-1]))
    variance.append(
        ((arr_view - store['mean'])**2).mean(axis=0)
    )
store['std'] = np.sqrt(np.stack(variance).mean(axis=0))
```

```python
store
```

```python
# Test
for i, batch in enumerate(ds):
    print(f"-- Progress test | {i}/{len(ds)} {(i*100/len(ds)):0.2f} %", end='\r')
    arr = batch['img']
    arr_view = arr.reshape((-1, arr.shape[-1]))
    arr_norm = (arr_view - store['mean'])/store['std']
    print(arr_norm.min(), arr_norm.max())
```
